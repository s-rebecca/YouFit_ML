```python
import pandas as pd
df = pd.read_csv("Activity_Dataset_V1.csv")
df.head()
```

```python
df1 = df[['vo2_max(%)']]
df1.head()
```

```python
df1.fillna(0, inplace=True)
```

```python
import numpy as np

df1['cardiorespiratory_fitness'] = np.where(df['vo2_max(%)'] < 35, 'less', 'good')
print(df1)
```

```python
df1.sample(20)
```

```python
from sklearn.preprocessing import MinMaxScaler
numeric_columns = df1.select_dtypes(include='number')
scaler = MinMaxScaler()
n_df1 = scaler.fit_transform(numeric_columns)
n_df1 = pd.DataFrame(n_df1, columns=numeric_columns.columns)
```

```python
n_df1['cardio-respiratory_fitness'] = df1['cardiorespiratory_fitness']
```

```python
n_df1.head()
```

```python
X = n_df1.drop('cardio-respiratory_fitness', axis=1)
y = n_df1['cardio-respiratory_fitness']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#from sklearn.naive_bayes import MultinomialNB
#nb_classifier = MultinomialNB()
#nb_classifier.fit(X_train, y_train)
from sklearn.naive_bayes import GaussianNB
nb_classifier = GaussianNB()
nb_classifier.fit(X_train, y_train)
```

```python
y_pred = nb_classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
